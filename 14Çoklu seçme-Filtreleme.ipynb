```python
import pandas as pd 
import numpy as np
```

```python
film = pd.read_csv("DataSets/imdbratings.csv")
film.head()
```

# seçme
Genel olarak 3 türlü seçme yapabiliriz;
1- köşeli parantez kullanmak.
2- loc methodunu kullanmak = 2 değer girilir ilk değer satırı, 2. değer sutunu gösterir. bu methodda etiket girilir
3- iloc methodunu kullanmak = yukardakiyle aynı fakat seçiceğimiz satır ve sutunların indexleri girilir.


```python
film["title"].head() #title sutununun ilk 5 satırı
```

```python
film[["title","genre","duration"]].head() #2 den çok ekrana yazdırmak istersek parantez içindwe parantez kullanılarak yapılır.
```

# LOC METHODU
Loc methodu için satır ve sutun etiketlerini girmek gerekir. ilk değer satır 2. sutunu gösterir.

```python
film.loc[[0,2,4],] #vırgulu koymayı unutma. loc methodunda ondan sonra bieş geliceğini anlamanda çok yardımcı olur.
```

```python
film.loc[0:5,]
```

```python
#Loc ile sutun işlemleri;
film.loc[0:5,"title"]
```

```python
#sutunları dilimlemek istersek şöyle yaparız;
film.loc[0:5,"title":"genre"]
```

```python
#loc methodunda sadece sutunları sıralayabliriz ;
film.loc[:,"title":"genre"]
```

```python
#LOC METHODU İLE SÜZME
film.loc[film.genre =="Crime",] #FİLM TURU CRİME OLAN SATIRLAR FİLTRELENDİ.
```

```python
film.loc[film.genre =="Crime","title"] #title sutunu crime değerine göre suzulmuş oldu. birden fazla değer allmak için köşeli parantez kullanıyoruz.
```

```python
film.loc[film.genre =="Crime","title":"duration",] #dilimlenmiş hali
```

```python
film.loc[film.genre =="Crime",["title","duration"]] #seçme bu şekilde.
```

# İLOC METHODU;
İlocun, loc methodundan farkı etiket yerine satır ve sutunların indexlerini girmemiz gerekir. İlk değer satır, 2. değer sutundur.

```python
film.iloc[:,0].head()
```

```python
film.columns #burdan indexleri bulabiliriz.
```

```python
film.iloc[:,[0,3]] #0. ve 3. indexe sahip sutunlar yazılmış oldu.
```

```python
film.iloc[:,0:3] #dilimleme yaparken parantez kalkıyo araya iki nokta geliyo
```

```python
film.iloc[0:5,0:3]#dilimleme
```

```python
film.iloc[0:5,:,]
```

# ÇOKLU FİLTRELEME SÜZME

```python
(film.duration>=200)&(film.genre =="Crime") # bu koşul boolean değer dönderir. eğer 
```

```python
film[(film.duration>=200)&(film.genre =="Crime")|(film.genre == "Action")] #böyle yaparsak da bu şekidle gösterir
```

```python
film[film.genre.isin(["Crime","Drama","Action"])] #türü parantez içindeki değerler süzülerek veri setinde ortaya çıkar.
```
