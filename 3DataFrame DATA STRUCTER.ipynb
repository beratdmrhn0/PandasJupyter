```python
#DATA FRAME VERİ YAPISI ÇOK BOYUTLUDUR. BİRDEN FAZLA SUTÜNDAN OLUŞUR
#Her bir sutünda sayısal,string,boolean gibi farklı veri tipleri bulunan dikdörtgen biçimindeki veri tablolarıdır.
import pandas as pd
```

```python
veri = {"isim":["Ali","Can","Efe","Ayşe","Buse","Alp","Nur"],
        "puan": [90, 80,85,75,95,60,65],
       "spor": ["Güreş","Futbol","Kayak","Yüzme","Tenis","Karate","Sörf"],
       "cinsiyet": ["E","E","E","K","K","E","K"]}
df = pd.DataFrame(veri)#Değerleri girerek bir data frame veri yapısı oluşturduk
df
```

```python
df.head() #ilk 5 veriyi gösterir.
```

```python
df.tail() #Son 5 satırı verir.
```

```python
df.tail(3) #sondan kaç tanesini görmek istiyosak parantez içine onu yazacaz.
#Parantez içine kaç tane görmek istiyosak yazabiliriz bu .head() ve .tail() içinde geçerlidir.
```

```python
df = pd.DataFrame(veri,columns =["isim","spor","cinsiyet","puan"]) 
df #Bu kod ile columns yazarak değişkenlerin yerlerini değiştirebiliriz.
```

```python
df = pd.DataFrame(veri,columns =["isim","spor","cinsiyet","puan","yas"])
df #sözlükte olmayan bir değer eklersek o satırın değerleri nan olarak yazar.
```

```python
df = pd.DataFrame(veri,columns =["isim","spor","cinsiyet","puan"],index = ["bir","iki","üç","dört","beş","altı","yedi"])
df #index =[] ile baştaki indexlere etiket atamış olduk.
```

```python
#veri setindeki belirli bir sutünü alabiliriz;
df["spor"]
```

```python
sutunlarım = ["isim","spor"] #birden fazla sutun almak için yeni bir değişken atamamız lazım
df[sutunlarım]
```

```python
#İstediğimiz sutüna df.ismi yazarakta ulaşabiliriz , örneğin;
df.spor
```

```python
#BELİRLİ SATIR SEÇMEK İÇİN .loc KULLANILIR ÖRNEĞİN;
df.loc[["bir","yedi"]]
```

```python
#Değerleri olmayan yaş sutünuna değer atayabiliriz örneğin;
df["yas"] = 18
df
```

```python
# Bir sutüna liste ya da dizi atamak istersek değerlerin uzunluğu DataFrame uzunluğu ile aynı olmalıdır.
deger = [17,18,19,17,18,20,18]
df["yas"] = deger
df
```

```python
df["gecti"]=df.puan>70 #puanı 70 den büyük olanlar için geçti isminde bir sutün oluştu ve değerleri altına yazıldı.
df
del df["gecti"] #Sutün silmek için del kullanılır.
df
```

```python
#Sözlük ilede DataFrame oluşturulabilir.
notlar = {"Mat" :{"Ali":85,"Efe":90,"Nur":95},
           "Fiz":{"Ali":90,"Efe":80,"Nur":75}}
puan = pd.DataFrame(notlar)
puan # iç içe sözlük içidne sözlük yapsıı dataframe ye dönüştürüldü.
#pandas modulu dıştakini sütün olarak, içtekinide satır olarak alır.
```

```python
#DataFramenin transpozu(Satır sutunları yer değiştirmesi)alınabilir.
puan.T
```

```python
#DataFramede satır ve sutunlara isim verilebilir.
puan.index.name ="isim" #indexlere isim atadı
puan.columns.name = "ders" #ders adlarına yani satıra ders atadı
puan
```

```python
#.values dataframedeki veri yapısını 2 boyutlu diziye dönüştürür.
puan.values
```

```python
#pandas modulundeki index niteliği veriyi değiştirilemez hale getirir.
indeks = puan.index
```
